```python
import numpy as np

def orthogonal_vector(v1, v2):
    # Convert input vectors to numpy arrays for easier operations
    v1 = np.array(v1)
    v2 = np.array(v2)

    # Normalize v1
    u1 = v1 / np.linalg.norm(v1)

    # Make v2 orthogonal to u1
    u2 = v2 - np.dot(v2, u1) * u1

    # Normalize u2
    u2 = u2 / np.linalg.norm(u2)

    # Generate a random vector
    r = np.random.randn(len(v1))

    # Make it orthogonal to both u1 and u2
    normal = r - np.dot(r, u1) * u1 - np.dot(r, u2) * u2

    # Normalize the resulting vector
    normal = normal / np.linalg.norm(normal)

    return normal

# Test the function
normal = [orthogonal_vector(v1, v2) for i in range(20)]

scores = {}
for i in range(len(vectors)):

  scores[others[i]] = np.mean([np.abs(np.dot(j, vectors[i])) for j in normal])
```
